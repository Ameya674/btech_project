```python
import pandas as pd
```

```python
df = pd.read_csv("./data/EDA_data.csv")
df
```

```python
df.describe()
```

```python
df.drop(['Unnamed: 0', 'I3 (W/m2)'], axis = 1, inplace = True)
df
```

```python
df.isnull().sum()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df
```

```python
df["Date"] = df["Date"].dt.round("H")
```

```python
df = df.drop_duplicates(subset='Date')
```

```python
# Set 'Date' column as the index
df.set_index('Date', inplace=True)

# Resample DataFrame to hourly frequency and fill missing rows with NaN
df = df.resample('1H').asfreq()

# Reset index to move 'Date' column back to a regular column
df.reset_index(inplace=True)
```

```python
df.interpolate(method='linear', inplace = True)
```

```python
df.tail(20)
```

```python
df.describe()
```

```python
df.to_csv("./data/clean_data.csv", index = False)
```
